```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as ax
```

```python
df= pd.read_csv(r"C:\Users\HP\Desktop\phyton.py\MATPLOTLIB.py\file\one.csv")
df.head(3)
```

```python
plt.style.use("fivethirtyeight")
```

```python
ages= df["Age"]
python= df["Python"]
java= df["JavaScript"]
```

```python
font= {"family" : "Times New Roman"}

plt.scatter(ages, python, marker="o", c=ages, cmap="CMRmap", edgecolor="black")

plt.title("Total Income", color="#C44040" ,fontdict=font)
plt.xlabel("Age", fontdict=font)
plt.ylabel("Software Language", fontdict=font)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
font={"family" : "Times New Roman"}
plt.scatter(ages, python, marker="o", c=ages, cmap="coolwarm", edgecolor="black", label="python")
plt.plot(ages, java, color="green", linewidth=1.5, label="java")


plt.title("Total Income", color="#EA920B" ,fontdict=font)
plt.xlabel("Age", fontdict=font)
plt.ylabel("Software Language", fontdict=font)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
font={"family" : "Times New Roman"}

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages, python, color="#B81265", linestyle="--", linewidth=1.75 ,label="python")
ax1.legend()
ax1.set_ylabel("Python", fontdict=font)
ax1.set_title("Total Income", color="#D15715", fontdict=font)

ax2.plot(ages, java, color="#D18D15", linewidth=1.75, label="java")
ax2.legend()
ax2.set_xlabel("Age", fontdict=font)
ax2.set_ylabel("Java", fontdict=font)

plt.tight_layout()
plt.show()
```

```python
font={"family" : "Times New Roman"}
plt.scatter(python, java, c=ages, cmap="PiYG", edgecolor="black")

cbar=plt.colorbar()
cbar.set_label("Age", fontdict=font)

plt.xticks(rotation=45, ha="right")
plt.xlabel("Python Income", fontdict=font)
plt.ylabel("JavaScript Income", fontdict=font)
plt.title("Total Income", color="darkblue", fontdict=font)
plt.legend()
plt.tight_layout()
plt.show()
```
