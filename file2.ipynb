```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv(r"C:\Users\HP\Desktop\phyton.py\MACHINE_LEARNING\df2.csv")
data.head(3)
```

```python
x= data["age"]
y= data["salary"]
z= data["expected_salary"]
```

```python
plt.style.use("fivethirtyeight")

plt.plot(x,y, color="#0B9F0B", marker="o", mec="black", linewidth=1.5, linestyle="dashed" ,label="salary")

plt.xlabel("AGE", color="#F2768B")
plt.ylabel("SALARY", color="#F2768B")
plt.title("Average Salary By Age", color= "#D3556A")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.style.use("fivethirtyeight")

plt.plot(x,z, color="#F06341",marker="o", ms=4, mec="k", linewidth=1.5, label="expected salary")

plt.xlabel("AGE", color="#F2768B")
plt.ylabel("EXPECTED SALARY", color="#F2768B")
plt.title("Average Expected Salary By Age", color= "#D3556A")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.style.use("fivethirtyeight")

plt.plot(x,y, color="#0B9F0B", marker="o", mec="black", linewidth=1.5, linestyle="dashed" ,label="current salary",)
plt.plot(x,z, color="#F06341", marker="o", ms=4, mec="k", linewidth=2, label="expected salary")

font= {"family": "serif"}
plt.xlabel("AGE", color="#F2768B", fontdict= font)
plt.ylabel("SALARY", color="#F2768B", fontdict= font)
plt.title("Average Salary By Age", color= "#D3556A", fontdict= font)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.style.use("fivethirtyeight")

plt.bar(x,y, color="#529244", label="current salary")
plt.plot(x,z, color="#F06341", marker="o", ms=3, mfc="black", linewidth=2, label="expected salary")

plt.xticks(x, fontsize=7)
plt.xlabel("AGE", color="#D1589C")
plt.ylabel("SALARY", color="#D1589C")
plt.title("Average Salary By Age", color="#D1589C")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn-v0_8-pastel")

plt.bar(x,y, color="#529244", label="current salary")
plt.scatter(x,z, color="#F06341", linewidth=2, label="expected salary")

plt.xticks(x, fontsize=7)
plt.xlabel("AGE", color="#D1589C")
plt.ylabel("SALARY", color="#D1589C")
plt.title("Average Salary By Age", color="#D1589C")
plt.tight_layout()
plt.legend()
plt.show()
```
