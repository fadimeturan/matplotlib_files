```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(r"C:\Users\HP\Desktop\phyton.py\MATPLOTLIB.py\file\one.csv")
df.head(3)
```

```python
h= df["HTML"].mean()
p= df["Python"].mean()
j= df["JavaScript"].mean()
```

```python
slices= [h, p, j]
labels= ["HTML", "PYTHON", "JAVASCRIPT"]
colors= ["#C83DF5", "#F88AD8", "#95EAB1"] 
explode= [0, 0.05, 0]

plt.pie(slices, labels=labels, explode=explode, colors=colors, wedgeprops={"edgecolor": "#552312"}, autopct="%1.0f%%" )
plt.title("Usages Of Languages")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```
