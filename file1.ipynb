```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv(r"C:\Users\HP\Desktop\phyton.py\cars.csv")
data= data.sort_values(by=["Volume"], ascending=True)           
data.head(5)
```

```python
Volume= data["Volume"]
Weight= data["Weight"]
```

```python
plt.figure(facecolor="#FFDFF8")

plt.scatter(Volume, Weight, color= "#CB1CA3", marker="o", edgecolor="black", label="wight/volume")

plt.xlabel("Car Volume")
plt.ylabel("Weight")
plt.title("Relation Between Car Volume and Weight")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.figure(facecolor="#FFDFF8")

plt.plot(Volume, Weight, color="#B80990", linewidth=1.75, label="weight/volume")

plt.xlabel("Car Volume")
plt.ylabel("Car Weight")
plt.title("Relation Between Car Volume and Weight")
plt.tight_layout()
plt.legend()
plt.show()
```
